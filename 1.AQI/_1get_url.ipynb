```
import os
import time
import requests
import sys
```

```
def retrieve_html():
  for year in range(2013,2019):
    for month in range(1,13):
      if(month<10):
        url = "https://en.tutiempo.net/climate/0{}-{}/ws-432950.html".format(month,year)
      else:
        url = "https://en.tutiempo.net/climate/{}-{}/ws-432950.html".format(month,year)
      source_text = requests.get(url)
      text_utf = source_text.text.encode('utf=8')
      if not os.path.exists(data_path+"Data/Html_Data/{}".format(year)):
        os.mkdir(data_path+"Data/Html_Data/{}".format(year))
      with open(data_path+"Data/Html_Data/{}/{}.html".format(year,month),"wb") as f:
        f.write(text_utf)
      sys.stdout.flush()
```

```
if __name__ =='__main__':
  start_time = time.time()
  retrieve_html()
  end_time = time.time()
  print("Retrieval time taken : {} ".format(end_time-start_time))
```
